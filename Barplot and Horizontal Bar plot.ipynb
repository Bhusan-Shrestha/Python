# Bar Plot

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
student={"Aayuscal Sedhai":55, "Hikmat Baniya":80, "Nishan Neupane":79, "Bhushan Shrestha":95}
```

```python
print(student)
```

```python
type(student)
```

```python
names=list(student.keys())
values=list(student.values())
```

```python
plt.bar(names,values)
plt.xticks(rotation=90)
plt.show()
```

# Adding titles and lables to the bar plot

```python
plt.bar(names,values)
plt.title("Bar Plot")
plt.xlabel("Name of a student")
plt.ylabel("Marks of a student")
plt.savefig('Bhushan.png')
plt.show
```
