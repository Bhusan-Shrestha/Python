# line Plot Basics

```python
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
x=np.arange(1,11)
```

```python
x
```

```python
y=2*x
```

```python
y
```

```python
plt.plot(x,y)
plt.show()
```

# Adding title and lables

```python
plt.plot(x,y)
plt.title("Line plot of x and y variables")
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.show()
```

# Changing line Aesthetics


```python
plt.plot(x,y,color="#808080",linestyle="--",linewidth=3)
plt.title("Line plot of x and y variable")
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.show()
```

# Adding two lines in the same plot

```python
y1=2*x
y2=3*x
```

```python
plt.plot(x,y1,color="g",linewidth=3)
plt.plot(x,y2,color="blue",linestyle=":",linewidth=5)
plt.grid(True,alpha=0.7)
```

```python
plt.subplot(1,2,1)
plt.plot(x,y1,color="g",linewidth=3)
plt.grid(True,alpha=0.7)
plt.subplot(1,2,2)
plt.plot(x,y2,color="blue",linestyle=":",linewidth=5)
plt.grid(True,alpha=0.7)
```

```python
plt.subplot(2,1,1)
plt.plot(x,y1,color="g",linewidth=3)
plt.grid(True,alpha=0.7)
plt.subplot(2,1,2)
plt.plot(x,y2,color="blue",linestyle=":",linewidth=5)
plt.grid(True,alpha=0.7)
```

```python
x=np.arange(2,101,2)
plt.figure(figsize=(20,15))

plt.subplot(2,3,1)
y1=np.exp(x)
plt.plot(x,y1,color="r",linestyle=":",linewidth=3)
plt.title("Line plot of y1=exp(x)")
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.grid(True,alpha=0.7)

plt.subplot(2,3,2)
y2=np.sin(x)
plt.plot(x,y2,color="b",linewidth=3)
plt.title("Line plot of y2=sin(x)")
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.grid(True,alpha=0.7)

plt.subplot(2,3,3)
y3=np.pi*np.cos(x)**2
plt.plot(x,y3,color="g",linewidth=3)
plt.title("Line plot of y3=pi*cos(x)**2")
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.grid(True,alpha=0.7)

plt.subplot(2,3,4)
y4=np.log(x)**2
plt.plot(x,y4,color="y",linewidth=3)
plt.title("Line plot of y4=log(x)**2")
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.grid(True,alpha=0.7)

plt.subplot(2,3,5)
y5=(x**2)+np.exp(-x)
plt.plot(x,y5,color="purple",linestyle="--",linewidth=3)
plt.title("Line plot of y5=(x)+exp(-x)")
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.grid(True,alpha=0.7)

plt.subplot(2,3,6)
y6=7*(x)**3+np.arctan(np.exp(x))
plt.plot(x,y6,color="black",linewidth=3)
plt.title("Line plot of y6=7*(x)**3+arctan*exp(x)")
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.grid(True,alpha=0.7)

plt.tight_layout(pad=3)
```
